```python
import pandas as pd
```

### Load and cleanup the Data

```python
data = pd.read_csv('top10s.csv',encoding='latin-1')
```

```python
data.head(20)
```

```python
# find out the shape of the data
data.shape
```

```python
# After load, show datasets info
data.info()
```

```python
# figure out the null values (if any)
data.isnull().sum()
```

```python
# droup out the duplicates observation (if any)
data = data.drop_duplicates()
```

```python
# check out the datasets' shape again. seems that there're no duplicated observations identified
data.shape
```

```python
# statistically describe the data
data.describe()
```

```python
# drop the unnamed columns since it's just an series of index+1, which we considered that they have no affects to the whole analysis
data = data.drop('Unnamed: 0',axis=1)
```

```python
# rename few of columns' name to make it makes more sense
data.rename(columns={'nrgy':'energy',
                     'dnce':'danceability',
                     'dB':'loudness',
                     'live':'liveliness',
                     'val':'valence',
                     'dur':'duration_sec',
                     'acous':'acousticness',
                     'spch':'speechniness',
                     'pop':'popularity'}
                     ,inplace=True)
```

```python
data.head()
```

### Statistical review of the data

#### find out most popular songs

```python
# find out the most popular song by sorting with descending order
data.sort_values(by=['popularity'],ascending=False)
```

```python
# find out the top 10 songs of 2010s
data.sort_values(by='popularity',ascending=False)[0:10]
```

```python
# find out the top 10 songs of 2018
data[data['year']==2018].sort_values(by=['popularity'],ascending=False)[0:10]
```

```python
# create list of dataframe object which contains top 10 songs of each year

annual10 = []
for year in data['year'].unique():
    df = data[data['year'] == year].sort_values(by=['popularity'],ascending=False)[0:10]
    annual10.append(df)
```

```python
print(annual10, type(annual10), type(annual10[0]))
```

#### Genre analysis

```python
# find out the unique genre for an entire datasets
data['top genre'].unique()
```

```python
# find out the most popularity by genre
data.groupby("top genre").agg(score=('popularity','mean'),
                              qty=('popularity','count')).sort_values(by='score',ascending=False)
```

### Artist analysis

```python
# find out top 20 on-chart artist
data.groupby('artist').agg(count=('popularity','count')).sort_values(by='count',ascending=False)[0:20]
```
